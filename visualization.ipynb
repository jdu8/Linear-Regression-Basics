## Importing Required Modules

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

Loading the dataset

```python
data=pd.read_csv("Ecommerce Customers.csv")
```

```python
data.head()
```

Cleaning the data

```python
data=data.replace("\n"," ",regex=True)
```

Some general information about the data

```python
data.head()
```

```python
data.describe()
```

```python
data.info()
```

```python
sns.pairplot(data)
```

```python
sns.jointplot(x='Time on App',y='Length of Membership',kind='hex',data=data)
```

```python
sns.histplot(data["Yearly Amount Spent"],kde=True,bins=50)
```

## Correlation Heatmap

```python
sns.heatmap(data.corr(),annot=True,cmap='viridis')
```

From the correlation plot we can see that Length of Membership has highest positive correlation with Yearly Amount Spent

## Regression Plots

```python
sns.lmplot(x='Length of Membership',y='Yearly Amount Spent',data=data)
```

```python
sns.lmplot(x='Time on App',y='Yearly Amount Spent',data=data)
```

```python
sns.lmplot(x='Avg. Session Length',y='Yearly Amount Spent',data=data)
```

```python
sns.lmplot(x='Time on Website',y='Yearly Amount Spent',data=data)
```
